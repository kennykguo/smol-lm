## streaming curation loop (template)
this uses hf datasets streaming mode and writes jsonl shards until the 30gb target is reached.


```python
target_gb = 0.1
target_bytes = target_gb * 1024**3
shard_size_bytes = 256 * 1024**2
```

```python
from datasets import load_dataset
from pathlib import Path
import json
import itertools

processed_dir = Path("../data/processed")
processed_dir.mkdir(parents=True, exist_ok=True)

def stream_dataset(spec):
    ds = load_dataset(spec["hf_id"], spec.get("config"), split=spec["split"], streaming=True)
    for row in ds:
        text = row.get(spec["text_field"])
        if text is None:
            text = row.get("text") or row.get("content") or row.get("code")
        if text is None:
            text = json.dumps(row, ensure_ascii=True)
        if text:
            yield text

def weighted_stream(specs):
    streams = []
    for spec in specs:
        if spec.get("skip_stream"):
            print("skipping", spec["name"], "for streaming")
            continue
        streams.append((spec, stream_dataset(spec)))
    while True:
        for spec, it in streams:
            repeats = max(1, int(spec["weight"] * 100))
            for _ in range(repeats):
                try:
                    yield next(it)
                except StopIteration:
                    break

def write_shards(text_iter, target_bytes, shard_size_bytes):
    shard_index = 0
    shard_bytes = 0
    total_bytes = 0
    shard_path = processed_dir / f"stage1_shard_{shard_index:04d}.jsonl"
    shard_file = open(shard_path, "w")
    for text in text_iter:
        record = {"text": text}
        line = json.dumps(record) + "\n"
        shard_file.write(line)
        shard_bytes += len(line.encode("utf-8"))
        total_bytes += len(line.encode("utf-8"))
        if shard_bytes >= shard_size_bytes:
            shard_file.close()
            shard_index += 1
            shard_bytes = 0
            shard_path = processed_dir / f"stage1_shard_{shard_index:04d}.jsonl"
            shard_file = open(shard_path, "w")
        if total_bytes >= target_bytes:
            break
    shard_file.close()
    print("total_bytes", total_bytes)

text_iter = weighted_stream(dataset_specs)
write_shards(text_iter, target_bytes, shard_size_bytes)
```

### manifest writer

```python
from pathlib import Path
import json
import hashlib

data_root = Path("../data")
manifest_path = data_root / "stages" / "stage1" / "manifest.jsonl"

def file_hash(path):
    h = hashlib.sha256()
    with open(path, "rb") as f:
        for chunk in iter(lambda: f.read(1024 * 1024), b""):
            h.update(chunk)
    return h.hexdigest()

def write_manifest(files, out_path):
    out_path.parent.mkdir(parents=True, exist_ok=True)
    with open(out_path, "w") as f:
        for p in files:
            record = {"path": str(p), "sha256": file_hash(p)}
            f.write(json.dumps(record) + "\n")

shards = sorted((data_root / "processed").glob("stage1_shard_*.jsonl"))
if shards:
    write_manifest(shards, manifest_path)
    print("wrote manifest", manifest_path)
else:
    print("no processed shards found; run curation first")
```

## token length histogram (small sample)
this helps spot extreme token lengths early.


```python
import tiktoken
import matplotlib.pyplot as plt
import json
from pathlib import Path

enc = tiktoken.get_encoding("gpt2")
lengths = []
processed_dir = Path("../data/processed")
shards = sorted(processed_dir.glob("stage1_shard_*.jsonl"))
for shard in shards[:1]:
    with open(shard, "r") as f:
        for i, line in zip(range(200), f):
            text = json.loads(line).get("text")
            if text is None:
                continue
            if not isinstance(text, str):
                text = str(text)
            lengths.append(len(enc.encode(text)))
if lengths:
    plt.hist(lengths, bins=30)
    plt.title("token lengths (sample)")
    plt.xlabel("tokens")
    plt.ylabel("count")
    plt.show()
else:
    print("no lengths to plot")
```
