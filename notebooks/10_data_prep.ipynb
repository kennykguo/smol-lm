# data prep

this notebook documents the local data layout and validates manifests.
use it to build stage-specific mixtures and snapshot metadata.


## data layout
- data/raw: source archives
- data/processed: cleaned text + tokenized shards
- data/stages: stage1, stage2, stage3 manifests
- data/arc: arc-agi datasets + synthetic grids
- data/metadata: manifests, hashes, filters

## snapshot workflow
- freeze stage manifests into a versioned dataset
- upload as smol-llm-stage1:v0.1 to hugging face
- pin snapshot tag in kaggle runs

## candidate sources
- fineweb-edu, openwebtext2, wikipedia, pg-19/public domain books
- the stack v2 (permissive subset), starcoder2data-edu
- gsm8k, mathqa, openwebmath
- arc-agi (public), arc-agi-2 public
- sudoku-extreme, maze-hard, synthetic grid transforms
